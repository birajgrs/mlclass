```python
import matplotlib.pyplot as plt
```

# 1. Creating Bar plot

```python
x=['Python','c/c++','c#','Java']
y=[86,73,77,83]
plt.title("Languages with popularities",fontsize=25)
plt.xlabel('Languages',fontsize=15)
plt.ylabel('Popularities level',fontsize=15)
c=['red','green','purple','pink']
plt.bar(x,y,width=0.3,color=c,align='edge')  #edge,center(default)
plt.show()
```

# 2. Pie plot

```python
x=[10,20,30,40]
y=['c/c++','java','python','php']
exp=[0.3,0.0,0.0,0.0]
c=['r','b','g','y']
plt.pie(x,labels=y,explode=exp,colors=c,autopct='%0.1f%%',shadow=True,radius=1.25)
plt.show()
```

# 3. Box plot and Whisker Plot 

```python
x=[10,20,30,40,50,60,70]
# plt.boxplot(x,notch=True)
# plt.boxplot(x,vert=False)
plt.boxplot(x,widths=0.3,labels=['Python'], patch_artist=True,showmeans=True)   #default width=0.2
plt.show()
```

# 4. Area plot with fill_between 

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 3, 2, 5]

plt.fill_between(x, y, color="skyblue", alpha=0.4)

plt.plot(x, y, color="Slateblue", alpha=0.6, linewidth=2)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Area Plot")

plt.show()
```
