# 1. Creating a Line Plot


```python

# initialize of data

x=[1,2,3,4,5]
y=[85,87,83,86,85]

# plotting the data
plt.plot(x,y)

# Adding the title
plt.title("Performance report")

#Assigning the labels
plt.xlabel("Semester")
plt.ylabel("Percentage of marks")

plt.show()
```

# 2. Creating a bar graph

```python

# initialize of data

x=[1,2,3,4,5]
y=[65,77,83,92,82]

# plotting the data
plt.bar(x,y)

# Adding the title
plt.title("Performance report")

#Assigning the labels
plt.xlabel("Semester")
plt.ylabel("Percentage of marks")

plt.show()
```

# 3. Creating Scatter graph

```python
# initialize of data

x=[1,2,3,4,5]
y=[65,77,83,92,82]

# plotting the data
plt.scatter(x,y)

# Adding the title
plt.title("Performance report")

#Assigning the labels
plt.xlabel("Semester")
plt.ylabel("Percentage of marks")

plt.show()
```

# 4. Creating Histogram plot

```python
from matplotlib import pyplot as plt

x=[1,3,5,4,7,8,6,9,8,2]
# This will plot a simple histogram
plt.hist(x,bins=[1,2,3,4,5,6,7,8,9])

# title to the plot
plt.title("Histogram")
# Adding legends
plt.legend("bar")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```

# 5. Creating Pie chart

```python
# data to display plots

x=[1,2,3,4]
e=[0.2,0,0,0]
# e=[0,0,0,0]

plt.pie(x,explode=e)

plt.title("Pie chart")
plt.show()
```

# 6. Creating 3D plot

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

plt.show()

```

# 7. Creating 3D line graph

```python
x=[1,2,3,4,5]
y=[1,4,8,16,25]
z=[1,8,27,64,125]
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.plot3D(z,y,x)

plt.show()
```
