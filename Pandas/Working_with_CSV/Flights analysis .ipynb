```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('flights.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
# plotting the trend of monthly passenger traffic for each airline
airlines = df.columns[2:]  
plt.figure(figsize=(12, 6))

for airline in airlines:
    plt.plot(df['month'], df[airline], label=airline)

plt.title('Monthly Passenger Traffic for Each Airline')
plt.xlabel('Month')
plt.ylabel('Passenger Count')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

# Plotting the overall monthly passenger traffic trend for all airlines combined
df['Total'] = df.iloc[:, 2:].sum(axis=1)  
plt.figure(figsize=(10, 6))
plt.plot(df['month'], df['Total'], color='blue', marker='o', linestyle='-')
plt.title('Overall Monthly Passenger Traffic for All Airlines')
plt.xlabel('Month')
plt.ylabel('Total Passenger Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
