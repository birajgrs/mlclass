```python
# Importing libraries

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Load DATASETS

df=pd.read_csv('bangalore house price prediction OHE-data.csv')
```

```python
df.head()
```

```python
# split data

x = df.drop('price', axis=1)
y = df['price']

print("Shape of x=", x.shape)
print("Shape of y=", y.shape)
```

```python
# applying train and test 

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=51)


print('Shape of X_train = ', x_train.shape)
print('Shape of y_train = ', y_train.shape)
print('Shape of X_test = ', x_test.shape)
print('Shape of y_test = ', y_test.shape)
```

```python
# Applying Feature Scaling

from sklearn.preprocessing import StandardScaler

sc=StandardScaler()

sc.fit(x_train)
x_train=sc.transform(x_train)
x_test=sc.transform(x_test)
```

# Linear Regression model

```python
from sklearn.linear_model import LinearRegression 

lr=LinearRegression()
```

```python
lr.fit(x_train,y_train )
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
# Predict the value of home

x_test[0,:]
```

```python
lr.predict([x_test[0, :]])
```

```python
lr.predict(x_test)
```

```python
y_test
```

```python
lr.score(x_test,y_test)
```

```python
#  Price Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Price vs. Total Square Feet

plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_sqft_int', y='price', data=df)
plt.title('Price vs. Total Square Feet')
plt.xlabel('Total Square Feet')
plt.ylabel('Price')
plt.show()
```

```python
# Price by Number of Bedrooms
plt.figure(figsize=(10, 6))
sns.boxplot(x='bhk', y='price', data=df)
plt.title('Price Distribution by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price')
plt.show()
```
