```python
# Importing Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# load in  data and save to a variable
df=sns.load_dataset("tips")
# df
df.head(10)
```

```python
df.tail()
```

# 1. Swarm plot

```python
# use swarmplot to visualize tip observations and amounts by day of the week

sns.swarmplot(x="day",y="tip",data=df)
sns.set_style("whitegrid")
plt.show()
```

```python
# visualize tip observations

sns.swarmplot(x=df['tip'])
sns.set_style('darkgrid')
plt.show()
```

```python
# color points by catefory 

gender_palette=['#a833ff','#ffaf33']
sns.swarmplot(x='day',y='tip',hue='sex',palette=gender_palette,data=df)
plt.show()
```

# 2. Violin plot

```python
sns.violinplot(x=df['tip'],color='gold')
plt.show()
```

```python
# Draw plot based on variable
sns.violinplot(x='sex', y='total_bill',data=df)
plt.show()
```

# 3. Facet Grids

```python
#draw facet grid based on tip variable

fg=sns.FacetGrid(df,col="time", row="sex")
fg=fg.map(plt.hist,'tip',color='tomato')
```

```python
fg = sns.FacetGrid(df, col="time", row="sex")
fg = fg.map(plt.scatter, 'total_bill', 'tip', color='floralwhite', edgecolor="hotpink")
```

```python
# plot by category 

x=sns.FacetGrid(df,col='time',hue='sex')
x=x.map(plt.scatter,'total_bill','tip')
x=x.add_legend()
```
