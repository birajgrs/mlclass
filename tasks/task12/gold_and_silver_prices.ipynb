```python
import urllib.request
import re
import csv
```

```python
url = "https://www.negja.org.np/index.php?page=rates"
response = urllib.request.urlopen(url)
webContent = response.read().decode('utf-8')

print(webContent)
```

```python
rows = re.findall(r'<tr>(.*?)</tr>', webContent, re.DOTALL)
data = []
```

```python
headers = re.findall(r'<th>(.*?)</th>', rows[0], re.DOTALL)
headers = [re.sub(r'\s+', ' ', header.strip()) for header in headers]
```

```python
for row in rows[1:]:  # Skip the header row
    cells = re.findall(r'<td.*?>(.*?)</td>', row, re.DOTALL)
    if cells:
        clean_cells = [re.sub(r'\s+', ' ', re.sub(r'<.*?>', '', cell).strip()) for cell in cells]
        data.append(clean_cells)
```

```python
with open('gold_rate_history.csv', 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    # Writing the headers
    csvwriter.writerow(headers)
    # Writing the data rows
    csvwriter.writerows(data)

print("Data has been successfully saved to gold_and_silver_rate_history.csv")
```
